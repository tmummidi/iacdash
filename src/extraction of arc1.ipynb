```python
import re

def extract_float_from_string(input_string):
    # Define a regular expression pattern to match numbers with optional dots
    pattern = r'\d*\.\d+|\d+\.\d*|\d+'

    # Use re.findall to find all matches in the input string
    matches = re.findall(pattern, input_string)[0]

    return "{:.4f}".format(float(matches))
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

acode = df['ARC Code'].copy()

def fetch_data_recursive(df, acode):
    # Check if all links are None or the DataFrame is empty
    if df['Link'].isna().all() or df.empty:
        return df

    all_dfs = []
    for i in range(len(df)):
        if pd.notna(df['Link'][i]):
            url = df['Link'][i]
            response = requests.get(url)
            soup = BeautifulSoup(response.text, "html.parser")

            all_tables = soup.find_all('table')
            right_table = soup.find('table')

            data = {
                'Link': [],
                'ARC Code': [],
                'Description': [],
                'Recommendations': [],
                'Recommended $ Savings': [],
                'Average Payback (yrs)': [],
                'Implementation Rate': [],
            }

            tab = right_table.findAll("tr")

            for i in range(len(tab)):
                cells = tab[i].findAll("td")
                if len(cells) < 5:
                    continue
                link = cells[0].find("a")["href"] if cells[0].find("a") else None
                arc_code_element = cells[1].text.strip()
                arc_code = extract_float_from_string(arc_code_element.split(' ')[0])
                description = cells[1].find("strong").next_sibling.strip()
                recommendations = cells[2].text.strip()
                savings = cells[3].text.strip()
                payback = cells[4].text.strip()
                imp_rate = cells[5].text.strip()
                if arc_code not in acode.values:
                    acode = acode.append(pd.Series([arc_code]))
                    print(arc_code)
                    data['Link'].append(link)
                    data['ARC Code'].append(arc_code)
                    data['Description'].append(description)
                    data['Recommendations'].append(recommendations)
                    data['Recommended $ Savings'].append(savings)
                    data['Average Payback (yrs)'].append(payback)
                    data['Implementation Rate'].append(imp_rate)

            df1 = pd.DataFrame(data)
            all_dfs.append(df1)

    all_dfs = pd.concat(all_dfs, ignore_index=True)
    return pd.concat([df, fetch_data_recursive(all_dfs, acode)], ignore_index=True)

# Call the recursive function with your initial DataFrame
result_df = fetch_data_recursive(df, acode)
print(result_df)
```
